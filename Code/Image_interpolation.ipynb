```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import math
```

```python
lr = cv2.imread('0829x8.png')
cv2_imshow(lr)
```

```python
HR = cv2.imread('0829.png')
cv2_imshow(HR)
```

```python
h = math.floor((lr.shape[1]*2)/4)*4
w = math.floor((lr.shape[0]*2)/4)*4
```

```python
cubic = cv2.imread('0829x8.png')
cubic = cv2.resize(cubic, (h,w), interpolation=cv2.INTER_CUBIC)
cv2_imshow(cubic)
```

```python
bilinear = cv2.imread('0829x8.png')
bilinear = cv2.resize(bilinear, (h,w), interpolation=cv2.INTER_LINEAR)
cv2_imshow(bilinear)
```

```python
nn = cv2.imread('0829x8.png')
nn = cv2.resize(nn, (h,w), interpolation=cv2.INTER_NEAREST)
cv2_imshow(nn)
```
